```python
import matplotlib.pyplot as plt
import numpy as np

# -------------------------
# Zones (marchés analysés)
# -------------------------
zones = ["Montréal", "Chine", "Toronto", "Dubaï", "Casablanca",
         "Brésil", "New York", "Mexique", "Suisse"]

# -------------------------
# Critères d’analyse
# -------------------------
criteres = ["Demande", "Salaire", "Maturité", "Concurrence", "Opportunités"]

# Exemple de données (0-10) pour chaque région selon chaque critère
donnees = [
    [8, 7, 8, 6, 7],  # Montréal
    [9, 8, 9, 9, 8],  # Chine
    [7, 7, 7, 6, 7],  # Toronto
    [6, 8, 7, 5, 6],  # Dubaï
    [5, 6, 5, 4, 5],  # Casablanca
    [8, 6, 7, 7, 8],  # Brésil
    [9, 9, 9, 8, 9],  # New York
    [7, 6, 6, 5, 6],  # Mexique
    [8, 8, 8, 7, 8],  # Suisse (valeurs d’exemple)
]

# -------------------------
# Préparation du graphique
# -------------------------
nb_var = len(criteres)

# Angles pour chaque axe
angles = np.linspace(0, 2 * np.pi, nb_var, endpoint=False).tolist()
angles += angles[:1]  # fermer le cercle

# Créer la figure
plt.figure(figsize=(10, 10))
ax = plt.subplot(111, polar=True)

# Tracer chaque zone
for i, zone in enumerate(zones):
    valeurs = donnees[i]
    valeurs += valeurs[:1]  # fermer le polygone
    ax.plot(angles, valeurs, label=zone, linewidth=1)
    ax.fill(angles, valeurs, alpha=0.1)

# Ajuster les étiquettes
ax.set_xticks(angles[:-1])
ax.set_xticklabels(criteres, fontsize=12)
ax.set_yticklabels([])  # enlever les valeurs radiales
plt.title("Cartographie du Marché - Data Scientist (9 zones)", fontsize=16, pad=20)

# Légende
plt.legend(bbox_to_anchor=(1.1, 1.05))
plt.show()
```
